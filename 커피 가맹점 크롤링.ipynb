```python
from selenium import webdriver
```

```python
# macos 사용, chromedriver 경로 지정
driver = webdriver.Chrome("/usr/local/bin/chromedriver")
driver.get("https://franchise.ftc.go.kr/user/extra/main/70/firHope/listBrand/jsp/LayOutPage.do")
```

```python
# select와 상호작용 위해 임포트
from selenium.webdriver.support.ui import Select
```

```python
# 피자 선택
select = Select(driver.find_element_by_id("selIndus"))
select.select_by_visible_text("커피")
select2 = Select(driver.find_element_by_id("selListType"))
select2.select_by_visible_text("가맹점 현황 정보")
```

```python
# 검색버튼 클릭
xpath = """//*[@id="frmSearch"]/input[5]"""
driver.find_element_by_xpath(xpath).click()
```

```python
# 테이블 읽기 및 csv 변환 위해 pandas 임포트
import pandas as pd
```

```python
# 테이블 찾고 내용 HTML 통해 읽기
xpath = """//*[@id="txt"]/div[3]/table"""
table_html = driver.find_element_by_xpath(xpath).get_attribute('outerHTML')
result = pd.read_html(table_html)
```

```python
result[0]
```

```python
df = result[0]
```

```python
# Get indexes where name column has value john
indexNames = df[(df['가맹점수'] == 0) & (df['신규개점'] == 0) & (df['계약종료'] == 0) &
               (df['계약해지'] == 0) & (df['명의변경'] == 0) & (df['가맹점평균매출액(매출액지수)'] == '0  (0.00)') & (df['가맹점면적(3.3㎡)당평균매출액(평균매출액지수)'] == '0  (0.00)')].index

# Delete these row indexes from dataFrame
df.drop(indexNames, inplace=True)

df
```

```python
# 읽은 내용 csv 엑셀 파일로 저장
df.to_csv('커피.csv', encoding='utf-8', index=False)
```
