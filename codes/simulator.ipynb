```python
inning = 0
# out = 3이면 회차 변경
out = 0
runner = 0
#각 팀 점수
home_score = 0
away_score = 0
#한 회에서 내는 점수
score = 0

"""
주자 규칙 정의
주자 無 = 0
주자 1루 = 1
주자 2루 = 2
주자 3루 = 3
주자 1,2루 = 4
주자 1,3루 = 5
주자 2,3루 = 6
주자 만루 = 7
"""

def one_hit(runner, score):
    # 주자 無, 주자 3루
    if runner == 0 or runner == 3:
        # 주자 1루로
        runner = 1
        # 주자 3루일 때는 득점
        score += 1
    # 주자 1루, 주자 1,3루
    if runner == 1 or runner == 5:
        # 주자 1,2루로
        runner = 4
        # 주자 1,3루일 때 득점
        if runner == 5:
            score += 1
    # 주자 2루, 주자 2,3루
    if runner == 2 or runner == 6:
        # 주자 1, 3루로
        runner = 5
        #주자 2,3루일경우 득점
        if runner == 6:
            score += 1
    # 주자 1,2루, 주자 만루
    if runner == 4 or runner == 7:
        # 주자 만루로
        runner = 7
        # 주자 만루일때는 득점
        if runner == 7:
            score += 1

    return runner, score

def two_hit(runner, score):
    # 주자 無, 주자 2루, 주자 2,3루
    if runner == 0 or runner == 2 or runner == 3 or runner == 6:
        # 주자 2루로
        runner = 2
        # 주자 2루일때, 주자 3루일 때, 득점 +1
        if runner == 2 or runner == 3:
            score += 1
        # 주자 2, 3루일때 득점 +2
        elif runner == 6:
            score += 2
    # 주자 1루, 주자 1,2루, 주자 만루
    if runner == 1 or runner == 4 or runner == 5 or runner == 7:
        # 주자 2,3루로
        runner = 6
        # 주자 1,2루일때 득점 +1
        if runner == 4 or runner == 5:
            score += 1
        # 주자 만루일때 득점 +2
        elif runner == 7:
            score += 2

    return runner, score

def three_hit(runner, score):
    if 1 <= runner and runner <= 3:
        score += 1
    elif 4 <= runner and runner <= 6:
        score += 2
    elif runner == 7:
        score += 3
    runner = 3
    return runner, score

def homerun(runner, score):
    # 주자 0명
    if runner == 0:
        score += 1
    # 주자 1명
    elif 1 <= runner and runner <= 3:
        score += 2
    # 주자 2명
    elif 4 <= runner and runner <= 6:
        score += 3
    # 주자 3명
    elif runner == 7:
        score += 4
    runner = 0
    return runner, score

def flyout(runner, score, out):
    # 주자 3루
    if runner == 3:
        # 주자 無
        runner = 0
    # 주자 1,3루
    elif runner == 5:
        # 주자 1루
        runner = 1
    # 주자 2,3루
    elif runner == 6:
        # 주자 2루
        runner = 2
    # 주자 만루
    elif runner == 7:
        # 주자 1,2루
        runner = 4
    #점수 +1, 아웃 +1
    score += 1
    out += 1
    return runner, score, out

def switch_result(result, runner, score, out):
    return {'볼넷':one_hit(runner, score),
            '사구':one_hit(runner, score),
            '안타':one_hit(runner, score),
            '2루타':two_hit(runner, score),
            '3루타':three_hit(runner, score),
            '홈런':homerun(runner, score),
            '희생플라이':flyout(runner, score, out),
            '아웃':out(out),
            '뜬공':out(out),
            '땅볼':out(out),
            '나가기':exit(0)}[result]

def base_out(out):
    out += 1
    return out

# 병살
# 1,2,3 → 0, 4 → 2, 5 → 3, 6 → 3, 7 → 6
def double_out(runner, out):
    if 1<= runner and runner <= 3:
        runner = 0
    elif runner == 4:
        runner = 2
    elif runner == 5:
        runner = 3
    elif runner == 6:
        runner = 3
    elif runner == 7:
        runner = 6
    out += 2
    return runner, out

while inning != 9.5:
    # 3아웃인 경우 체인지
    while out != 3:
        result = input("결과 : ")
        print("주자 : %d, 아웃 : %d, 점수 : %d" %(runner, score, out))
    inning += 0.5
    out = 0
    runner = 0
    # 점수 기록
        # 이닝이 0으로 나눠떨어지면 홈팀
    if inning % 1 == 0:
        home_score = score
    # 이닝이 0으로 나눠떨어지지 않으면 어웨이
    else:
        away_score = score

if home_score > away_score:
    print("홈 승리")
elif home_score == away_score:
    print("무승부")
else:
    print("어웨이 승리")
```
