```python
from selenium import webdriver
import bs4
from tqdm import tqdm
import requests
import time
import pandas as pd
import numpy as np
import os
import csv
```

```python
team_list = ["두산", "LG", "키움", "SK", "KT", "한화", "삼성", "KIA", "롯데", "NC"]
year_list = list(range(2015, 2021))

statiz = "http://www.statiz.co.kr/"
chrome_dir = r'C:\Users\wkdgy\Documents\Jupyter\chromedriver.exe'
driver = webdriver.Chrome(chrome_dir)
driver.implicitly_wait(5) #크롬 설정
```

```python
def Refactoring():
    for team in team_list:
        DIR = "./KBO/"+team+"/"
        print(team, " 리팩토링 중 ...")
        
        file_list = os.listdir(DIR)   #파일 리스트 불러옴
        
        for file in file_list:
            
            result = []
            
            player = open(DIR + file, 'r')
            print(DIR+file)
            lines = csv.reader(player)
            
            for line in lines:
                if(line[1] == "연도"):
                    continue
                result.append(line[1:31])
            df = pd.DataFrame(result, columns=["연도", "팀", "나이", "포지션", "출장", "타석", "타수","득점"," 안타", "2루타", "3루타", "홈런", "루타", "타점", "도루", "도실", "볼넷", "사구", "고4", "삼진", "병살", "희타", "희비","타율", "출루","장타", "OPS", "wOBA", "wRC+", "WRA"])
            df.to_csv("./KBO/"+team+"/결과/"+file, encoding='cp949')
            print(file, " 재생성 중...")
```
