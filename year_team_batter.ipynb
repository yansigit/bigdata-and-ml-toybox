```python
from selenium import webdriver
import bs4
import requests
import time
import os
import csv
import pandas as pdyear
import numpy as np
```

```python
team_list = ["두산", "LG", "히어로즈", "SK", "KT", "한화", "삼성", "KIA", "롯데", "NC"]
year_list = list(range(2015, 2021))

statiz = "http://www.statiz.co.kr/"
chrome_dir = r'C:\Users\wkdgy\Documents\Jupyter\chromedriver.exe'
driver = webdriver.Chrome(chrome_dir)
driver.implicitly_wait(5) #크롬 설정
```

```python
for year in year_list:
    year = str(year)
    
    print(year, "년도 시작...")
    
    for team in team_list:
        print(team, " 프린팅 중...")
        url = "http://www.statiz.co.kr/stat_at.php?opt=0&sopt=0&re=0&ys="+year+"&ye="+year+"&se=0&te="+team+"&tm=&ty=2014&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR_ALL_ADJ&o2=TPA&de=1&lr=0&tr=&cv=&ml=1&sn=30&si=&cn="
        driver.get(url)

        source = driver.page_source
        tables = pd.read_html(source)

        tables[1].to_csv("./KBO/"+year+"/"+team+".csv", encoding='cp949')
```

```python
for year in year_list:
    year = str(year)
    print(year, "년도 재설정 시작...")
    Dir = "./KBO/"+year+"/"
    file_list = os.listdir(Dir)
    
    for file in file_list:
        result = []
        
        team = open(Dir + file, 'r')
        lines = csv.reader(team)
        
        for line in lines:
            if(line[1] == "순"):
                continue
            if(line[3][2:] == "P"):
                continue
            result.append(line[2:30])
        df = pd.DataFrame(result, columns = ["이름", "팀", "WAR", "출장", "타석", "타수", "득점", "안타", "2루타", "3루타", "홈런", "루타", "타점", "도루", "도실", "볼넷", "사구", "고4", "삼진", "병살", "희타", "희비", "타율", "출루율", "장타율", "OPS", "wOBA", "wRC+"])
        df.to_csv(Dir+file, encoding='cp949')
```
