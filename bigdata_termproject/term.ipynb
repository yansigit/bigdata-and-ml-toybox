```python
import pandas as pd
from IPython.core.display import display
```

```python
df1 = pd.read_csv("2019_관객수_및_매출액_월별.csv")
df2 = pd.read_csv("2020_관객수_및_매출액_월별.csv")
df3 = pd.read_csv("연도별_관객수_및_매출액.csv")
# display(df1)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt

# 폰트 설정
mpl.rc('font', family='NanumSquare')
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
def display_graph(df, title):
    if pd.api.types.is_string_dtype(df['전체영화매출액']):
        df['전체영화매출액'] = df['전체영화매출액'].str.replace(pat=',', repl='', regex=False)
        df['전체영화매출액'] = pd.to_numeric(df['전체영화매출액'])
    graph = df['전체영화매출액'][:-1].plot(kind='bar',x='년월',y='전체영화매출액',title=title)
    graph.set_xticklabels(df['년월'][:-1])
```

```python
display_graph(df1, "2019_관객수_및_매출액_월별.csv")
```

```python
display_graph(df2, "2020_관객수_및_매출액_월별.csv")
```

```python
display_graph(df3, "연도별_관객수_및_매출액.csv")
```
