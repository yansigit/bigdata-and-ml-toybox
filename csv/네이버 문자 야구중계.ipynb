```python
from time import sleep
from selenium import webdriver
import re
import numpy as np
import pandas as pd
import os
from IPython.core.display import display
```

```python
# macos 사용, chromedriver 경로 지정
driver = webdriver.Chrome("/usr/local/bin/chromedriver")

season = "2020"; date= "1017"
teamId = {"한화":'HH',"기아":'HT',"삼성":'SS',"LG":'LG',"KT":'KT',"두산":'OB',"NC":'NC',"SK":'SK',"롯데":'LT',"키움":'WO'}
gameId = season + date + teamId['두산'] + teamId['키움'] + "0" + season
url = "https://sports.news.naver.com/gameCenter/textRelay.nhn?gameId=" + gameId + "&category=kbo"

driver.implicitly_wait(5)
```

```python
gameOnGoing = True
while gameOnGoing:
    driver.get(url)
    driver.find_element_by_id("inning_tab_all").click()

    current_away_player = driver.find_element_by_xpath('//*[@id="away_player_box"]/div/dl/dt/a[1]').get_attribute("innerText")
    current_home_player = driver.find_element_by_xpath('//*[@id="home_player_box"]/div/dl/dt/a').get_attribute("innerText")

    _text = driver.find_element_by_id("relay_text").get_attribute('innerText')
    _text = re.sub('- (\d)구', r'\1', _text)
    _text = _text.replace('\n\n','\n')

    print('어웨이 플레이어 = ' + current_away_player)
    print('홈 플레이어 = ' + current_home_player)
    print(_text)

    if '승리투수' in _text:
        gameOnGoing = False
    else:
        sleep(10*1000)
```
